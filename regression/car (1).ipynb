```python
x=df.drop(columns='price',axis=1)
y=df['price']
```

```python
x
```

encoding   the whole data at once using one hot encoding

```python
# x_one_encode=pd.get_dummies(x)
```

```python
x.columns
```

```python
x_one_encode=pd.get_dummies(x,columns=['model','transmission','fuelType'],drop_first=True)
```

after encoding we got multiple columns of categorical features seperately

```python
x_one_encode
```

```python
x_one_encode=x_one_encode.astype(int)
```

converting the whole voolean data that came when doing onehot  encoding into integer

```python
x_one_encode  ##one hot  encoded data
```

```python

from sklearn.preprocessing import LabelEncoder
encode =LabelEncoder()

col=['model','transmission','fuelType']

xlavel=x

for i in col:
  xlavel[i]=encode.fit_transform(xlavel[i])
```

using lavel encoding from livrary from sklearn
from this encoding  various column will not ve created  instead of that encoding is done within the same column

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

numerical_col=['year','mileage','tax','mpg']

# x_one_encode[numerical_col]=scaler.fit_transform(x_one_encode[numerical_col])
x_one_encode[numerical_col]=scaler.fit_transform(x_one_encode[numerical_col])

```

```python
x_one_encode
```

```python
xlavel.columns
```

```python
xlavel[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']]=scaler.fit_transform(xlavel[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']])
```

making model applying llinear regression for one hot encoded data

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression


X_train, X_test, y_train, y_test = train_test_split(x_one_encode, y, test_size=0.33, random_state=42)

```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
```

```python
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import  r2_score
r2=r2_score(y_test,y_predict)
r2


p=X_test.shape[1]
n=X_test.shape[0]


adjusted_r2=1-((1-r2)*(n-1)/(n-p-1))
adjusted_r2
```
